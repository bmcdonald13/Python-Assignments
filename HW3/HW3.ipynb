## Import Data Step
This includes the introduction of the pandas and matplot libraries. 
The data is provided with a header and displayed. 



```python
import pandas as pd 
import matplotlib.pyplot as plt

Data = pd.read_csv('HW3Data.csv',header = None)

Data
```

## Training/Testing Split Stage
During this stage, the labels and features are ascertained as individual datasets.
After this, sckikit model selection tool is used to import the splittin process. A random state of 100 is used as we well as an 80/20 Training/Test.

```python
Labels = Data.loc[:,0].values
Features = Data.loc[:,1:].values

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = \
    train_test_split(Features, Labels, 
                     test_size=0.20,
                     stratify=Labels,
                     random_state=6938)
```

## Recursive Feature Elimation through Cross Validation Stage
Within this stage, SVC and RFECV are utilized to obtain the optimal number of features. 
After this, the cross validation score is plotted against the number of features selected

```python
from sklearn.svm import SVC
from sklearn.feature_selection import RFECV

svc = SVC(kernel="linear")
rfecv = RFECV(estimator=svc, step=1, cv=10,
              scoring='accuracy')
rfecv.fit(X_train, y_train)

print("Optimal number of features : %d" % rfecv.n_features_)
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score (nb of correct classifications)")
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
plt.show()
```

## SVC Score

```python
svc.fit(X_train[:,rfecv.support_], y_train)
svc.score(X_test[:,rfecv.support_], y_test)
```
